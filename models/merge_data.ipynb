```python
import pandas as pd

# 1. Load both datasets (Make sure they are in your 'data' folder)
df_original = pd.read_csv('/Users/sp/Desktop/Fitness_App_Project/data/gym_members_exercise_tracking.csv')
df_synthetic = pd.read_csv('/Users/sp/Desktop/Fitness_App_Project/data/gym_members_exercise_tracking_synthetic_data.csv')

# 2. Stack them directly on top of each other
df_combined = pd.concat([df_original, df_synthetic], ignore_index=True)

# 3. Save the new master dataset in the exact same folder
df_combined.to_csv('gym_members_master.csv', index=False)

print(f"🔥 Success! Master dataset created with {len(df_combined)} rows.")
```
